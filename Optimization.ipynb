# Main Algorithm 

```python
# 'i' is the index for planes (1 to n_p)
# 'j','k' is the index for cities (1 to n_c)
# 'd' is the index for days (1 to n_d), here (0 to 6)


# 'j(d)' is the node for city j on day d
# 'j(d-1)k(d)r' is the edge connecting city j on day d-1 to city k on day d via route r

# 'G(k(d))' is the set of nodes connnected to k(d) 
# 'F(k(d))' is the set of nodes k(d) is connnected to

# 'L' is set of all edges
# 'N' is set of all nodes

# 'g(i,j)' is the cost for plane i at city j
# 'p(j)' is the number of planes that can take maintenance at city j
# 'y(i)'=1, is the number of days until the due date for maintenance

# 'w(i,j,d)'=1, if maintenance taken, else 0
# 'C(i,j(d-1)k(d)r)' is the cost for plane i via j(d-1)k(d)r edge
# 'x(i,j(d-1)k(d)r)'=1, if plane i is taken via j(d-1)k(d)r edge, else 0
```

```python
#This is the number of cities in the network
n_c = len(df1['Origin City'].unique())

#This is the number of planes in the fleet
n_p = len(df2['Aircraft'])

#This is the number of days of planning horizon
n_d = 7 #week
```

```python
#Print the values of all variable
print('no. of cities',n_c)
print('no. of planes',n_p)
print('no. of days',n_d)
```

```python
'''
create_network function initializes a 3D array for storing origin city as rows,
destination city as columns, days as depth, and existance of route as values.
This gives the possible routes for each day. 

We initially used dictionary as a form of data structure for storing the 
possible, and final routes. However, Dictionary uses associative array data 
structure that is of O(N) space complexity and a single add takes amortized O(1) 
time. Time complexity changes to O(N) when the underlying data structure needs 
to grow or shrink.
However, handling with arrays proves to be more convinient along with easy 
itteration and adding, changing values. Space complexity of an array is the same
as a dictionary, O(N). Time complexity depends on the size of the array. For 
example, an array of size N will require a time complexity of O(N).
'''
def create_network(df_new):
  A=np.zeros([n_c,n_c,n_d])
  for i in range(n_d):
    df=df_new[df_new['Day of Week']==i].copy()
    df=df.reset_index(drop=True)
    for i,li in enumerate(df['Origin City']):
      A[df['Origin City'].iloc[i]-1,df['Destination City'].iloc[i]-1]=1
  return A

network_array = create_network(df1)
print(network_array.shape)
```

```python
'''
dfs function outputs a cyclicity, cost and route path for a week's schedule. It 
performs recursive depth first search optimization algorithm to minimize the 
cyclic cost of the airplane's schedule.
Here the cost for scheduling a plane on any route is considered homogenous, with
a constant value of 0.
'''
INF = 10e10

def dfs(n_p,ori,k,d,net_arr):
  if d==6:
    path=[]
    if net_arr[ori][k][d]==1:
      path.append(k)
      cost=0
      return True,cost,path
    else:
      return False,INF,path
  else:
    curr_cost = INF
    is_cyclic = False
    curr_path = []
    for i in range(15):
      ind = i
      if net_arr[k,i,d]:
        #print(network_array[k,i,d])
        p_cyclic, p_cost, p_path = dfs(n_p,ori,i,d+1,net_arr)
        if p_cyclic == True:
          is_cyclic = True
          if p_cost <= curr_cost:
            ind = i
            curr_cost = p_cost
            curr_path = []
            curr_path = p_path
            curr_path.append(i)

    if d==4 and is_cyclic==True:
      curr_cost += g[n_p,ind-1]
    return is_cyclic, curr_cost , curr_path
```

```python
# this is the final output of the heuristics
planning_schedule = []
objective_function = 0
```

```python
'''
The function runs for some fixed number of epochs and providing an optimal 
cyclic path for each plane. Along with the comparision to a randomdly generated
path for the planes. 
'''
import time

#flowchart
# number of iterations
num_iter = 5000
for iter in range(num_iter):
  start = time.time()
  network_array_copy_1 = network_array
  final_array_copy_1 = final_array
  objective_func_1 = 0
  planning_schedule_1=[]
  for n in range(0,n_p):
    cyclic=False
    k=1
    while cyclic!=True:
      d = 0 
      while np.sum(network_array_copy_1[k,:,d])==0:
        k = k+1
      cyclic,cost,path = dfs(n,k,k,d,network_array_copy_1)
      if(cyclic):
        planning_schedule_1.append(path)
        for i in range(len(path)-1):
          final_array_copy_1[n,path[i],path[i+1],d]=1
          network_array_copy_1[path[i],path[i+1],d]=0
          d+=1
    objective_func_1 += cost
  network_array_copy_2 = network_array
  final_array_copy_2 = final_array
  objective_func_2 = 0
  planning_schedule_2=[]
  temp = np.array([i for i in range(n_p)])
  random.shuffle(temp)
  for n in temp:
    cyclic=False
    k=1
    while cyclic!=True:
      d = 0 
      while np.sum(network_array_copy_2[k,:,d])==0:
        k = k+1
      cyclic,cost,path = dfs(n,k,k,d,network_array_copy_2)
      if(cyclic):
        planning_schedule_2.append(path)
        for i in range(len(path)-1):
          final_array_copy_2[n,path[i],path[i+1],d]=1
          network_array_copy_2[path[i],path[i+1],d]=0
          d+=1
    objective_func_2 += cost
  if objective_func_1 <= objective_func_2 :
    objective_function = objective_func_1
    final_array = final_array_copy_1
    network_array = network_array_copy_1
    planning_schedule = planning_schedule_1
  else:
    objective_function = objective_func_2
    planning_schedule = planning_schedule_2
    final_array = final_array_copy_2
    network_array = network_array_copy_2
  end =time.time()
  print('Ran successfully for iteration ',i,"and time taken till this iteration is ",(end-start))
```
